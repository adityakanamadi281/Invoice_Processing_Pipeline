```python
from pydantic import BaseModel
from typing import List

class LineItem(BaseModel):
    description: str
    quantity: float
    unit_price: float
    line_total: float


class Invoice(BaseModel):
    Invoice_Number: str
    Invoice_Date: str
    Vendor_Name: str
    Ship_To: str
    Cust_Id: str
    Order_Date: str
    Invoice_Total: float
    items: List[LineItem]
```

```python
import pdfplumber
import re


def extract_text_from_pdf(pdf_path):

    text = ""

    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            page_text = page.extract_text()
            if page_text:
                text += page_text + "\n"

    return text
```

```python
def extract_invoice_fields(text):

    header_match = re.search(
        r'NET\s+\d+\s+DAYS\s+(\d+)\s+(\d+)\s+(\d{2}/\d{2}/\d{4})',
        text
    )

    invoice_match = re.search(
        r'INVOICE DATE\s*\n(\d{2}/\d{2}/\d{4})',
        text
    )

    total_match = re.search(
        r'INVOICE TOTAL.*?\$?\s*([\d]+\.\d+)',
        text,
        re.S
    )

    return {
        "Invoice_Number": header_match.group(1) if header_match else "",
        "Cust_Id": header_match.group(2) if header_match else "",
        "Order_Date": header_match.group(3) if header_match else "",
        "Invoice_Date": invoice_match.group(1) if invoice_match else "",
        "Vendor_Name": "Pacific Food Importers",
        "Ship_To": "Seattle WA",
        "Invoice_Total": float(total_match.group(1)) if total_match else 0.0
    }
```

```python
def extract_items(text):

    items = []

    lines = text.split("\n")

    for line in lines:

        parts = line.split()

        if len(parts) > 8 and parts[0].isdigit():

            try:
                quantity = float(parts[1])
                unit_price = float(parts[-3])
                line_total = float(parts[-1])

                description = " ".join(parts[4:-4])

                item = LineItem(
                    description=description,
                    quantity=quantity,
                    unit_price=unit_price,
                    line_total=line_total
                )

                items.append(item)

            except:
                pass

    return items
```

```python
import os
import json

pdf_folder = "/content/invoices"   # folder containing PDFs

all_invoices = []

for file in os.listdir(pdf_folder):

    if file.endswith(".pdf"):

        pdf_path = os.path.join(pdf_folder, file)

        print("Processing:", file)

        text = extract_text_from_pdf(pdf_path)

        invoice_data = extract_invoice_fields(text)

        items = extract_items(text)

        invoice = Invoice(
            **invoice_data,
            items=items
        )

        all_invoices.append(invoice.model_dump())
```

```python
with open("invoices_output.json", "w") as f:
    json.dump(all_invoices, f, indent=2)

print("All invoices saved to invoices_output.json")
```

```python
import csv

invoice_rows = []
lineitem_rows = []

for inv in all_invoices:

    # skip empty invoices
    if not inv["Invoice_Number"]:
        continue

    invoice_rows.append({
        "Invoice_Number": inv["Invoice_Number"],
        "Invoice_Date": inv["Invoice_Date"],
        "Vendor_Name": inv["Vendor_Name"],
        "Ship_To": inv["Ship_To"],
        "Cust_Id": inv["Cust_Id"],
        "Order_Date": inv["Order_Date"],
        "Invoice_Total": inv["Invoice_Total"]
    })

    for item in inv["items"]:

        lineitem_rows.append({
            "Invoice_Number": inv["Invoice_Number"],
            "description": item["description"],
            "quantity": item["quantity"],
            "unit_price": item["unit_price"],
            "line_total": item["line_total"]
        })
```
